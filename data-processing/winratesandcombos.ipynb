```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./matchs200kfull.csv",header=None)
df.columns=["uid","rank","gid","player","name","result","turn","cardid"]
```

```python
df.head()
```

```python
df["cardwin"]=((df["player"]=="me") & (df["result"]=="win")) | ((df["player"]=="opponent") & (df["result"]=="loss"))
df["w"]=1
df["rankrange"] = np.round(df["rank"]/5)
df["mano"]=df["gid"].apply(str)+" _ "+df["player"]
```

```python
dfwinrates=df[["name","cardwin","w"]].groupby(["name"]).sum()
```

```python
dfwinrates["winrates"]=dfwinrates["cardwin"]/dfwinrates["w"]
```

```python
dfwinrates["wp"]=dfwinrates["w"]/df["w"].sum()
```

```python
dfwinrates.sort_values("wp",inplace=True)
dfwinrates
```

```python
df["deckidturn"]=df["gid"].apply(lambda i: str(i)+'_')+df["turn"].apply(lambda i: str(i)+'_')+df["player"]
```

```python
dfsmall=df.iloc[1:500000,:]
mj=pd.merge(df,df,on='deckidturn')
mj.head()
```

```python
combos=mj[["name_x","name_y","cardwin_x","w_x","turn_x","turn_y"]].groupby(["name_x","name_y"]).sum()
combos
```

```python
combos.reset_index(inplace=True)
combosdf=combos.loc[(combos["w_x"]>200) & (combos["name_x"]!=combos["name_y"]),:]
```

```python
combosdf.loc[:,("winrates")]=combosdf["cardwin_x"]/combosdf["w_x"]
combosdfs=combosdf.sort_values("winrates",ascending=False)
combos_finals=combosdfs.iloc[range(0,combosdfs.shape[0],2),:]
combos_finals=combos_finals.loc[combos_finals["winrates"]>0.6,:]
```

```python
combos_finals
```

```python
dfwinrates.reset_index(inplace=True)
dfwinrates_x=dfwinrates[["name","winrates"]]
dfwinrates_x.columns=["name_x","winrates_x"]
dfwinrates_y=dfwinrates[["name","winrates"]]
dfwinrates_y.columns=["name_y","winrates_y"]
```

```python
combos_finalsm=pd.merge(combos_finals,dfwinrates_x,on='name_x')
combos_finalsm=pd.merge(combos_finalsm,dfwinrates_y,on='name_y')
combos_finalsm.loc[:,"combowin"]=np.log(combos_finalsm["winrates"]/(combos_finalsm["winrates_x"]*combos_finalsm["winrates_y"]))
ccf=combos_finalsm.sort_values("combowin",ascending=False)
ccf["meanturn"]=ccf["turn_x"]/ccf["w_x"]
```
